```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,9

import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('C:\\Users\\yusuper\\Desktop\\win_7\\pyda\\L10\\HW\\tmdb_5000_movies.csv')
movies
print(movies.shape)
```

```python
movies = movies.dropna()
print(movies.shape)
```

```python
movies.head()
```

```python
plt.barh(movies['popularity'], movies['runtime'])
plt.xlabel('popularity')
plt.ylabel('runtime')
plt.grid()
# график показывает  зависимость популярности фильма от его продолжительности
```

```python
movies.plot('vote_count', 'vote_average')
# график показывает зависимость величины оценки от их количества
```

```python
movies_3rd_graph =  movies.groupby('original_language')[['popularity']].mean()
movies_3rd_graph.plot(kind='pie', y = 'popularity' )
# график показывает зависимость популярности фильма от языка. Редкий случай, когда с натяжкой можно сказать, 
# что pay chart хоть сколько-нибудь информативен
```

```python
import seaborn as sns
```

```python
sns_plot = sns.pairplot(
    movies[['budget', 'revenue', 'vote_average']])
sns_plot
# строит пары зависимостей бюджета, сборов и средних оценок
```

```python
sns.jointplot(x='budget', y='vote_average', 
              data=movies, kind='scatter')
# зависимость бюджета и средней оценки
```
